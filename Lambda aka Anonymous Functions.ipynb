## example

`(lambda x, y : x * y)`


## Invoke lambda functions

To run the function use parentheses.

`(lambda x, y : x * y)()`

```python
(lambda x=5, y=10: x*y)()
```

```python
def mult(i, j=6):
    return i * j

mult(3)
```

```python
## is the same as
mult = (lambda i,j=6 : i * j)

mult(3)
```
