## Copying nested data

### shallow copy

When you copy a list with nested mutables rather than cloning it, the outer list is a non-aliased copy and won't change if the original changes.

BUT, any nested mutables are still aliased and will reflect any changes to the original nested mutables.

```python
orig = [['cats', 'dogs'], ['fox', 'deer']]
clone = orig
copy = orig[:]
```

```python
print(orig)
print(clone)
print(copy)
```

```python
print(orig == clone)
print(orig is clone)
print()
print(orig == copy)
print(orig is copy)
```

```python
orig.append(['tiger', 'lion'])
print(orig)
print(clone)
print(copy)
```

```python
print(orig[0] == clone[0])
print(orig[0] is clone[0])
print()
print(orig[0] == copy[0])
print(orig[0] is copy[0])
```

```python
orig[0].append('rabbit')
print(orig)
print(clone)
print(copy)
```

### deep copy

Use iteration to deep copy a list of nested mutables to ensure you do not alias those mutables

```python
orig = [['pots', 'pans'], ['cups', 'saucers']]
copy = []
for item in orig:
    copy.append(item[:])
    
print(orig)
print(copy)
```

```python
print(orig == copy)
print(orig is copy)
print()
print(orig[1] == copy[1])
print(orig[1] is copy[1])
```

```python
orig.append(['salt', 'pepper'])
orig[1].append('plates')

print(orig)
print(copy)
```

### copy module

use the copy module to do shallow and deep copies easily

* `copy.copy(x)` shallow copies x
* `copy.deepcopy(x)` deep copies x

```python
# import copy module
import copy
```

```python
orig = [['shoes', 'sneakers'], ['coats', 'jackets']]
s_copy = copy.copy(orig)
d_copy = copy.deepcopy(orig)
print(orig)
print(s_copy)
print(d_copy)
```

```python
orig.append(['shirt', 'pants'])
orig[0].append('heels')

print(orig)
print(s_copy)
print(d_copy)
```
